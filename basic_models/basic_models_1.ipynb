```python
data = np.array([[1,2,3,4],[12,28,32,56]])
```

```python
# asginar x e y
x = data[0]
y = data[1]

# calcular los estadisticos necesarios
x_mean = x.mean()
y_mean = y.mean()

# asignar b0 y b1
b1 = np.sum([(x_i-x_mean) * (y_i-y_mean) for x_i,y_i in zip(x,y)])/np.sum([(x_i - x_mean)**2 for x_i in x])
b0 = y_mean - b1 * x_mean

error = np.mean([(y_i - (b0+x_i*b1))**2 for x_i,y_i in zip(x,y)])

# graficar la recta
plt.scatter(x,y)
plt.plot(x, b0 + x*b1)
plt.show()

print("\n",b0,b1)
```

```python
x_minmax = np.array([(xi - x.min()) / (x.max() - x.min()) for xi in x])
y_minmax = np.array([(yi - y.min()) / (y.max() - y.min()) for yi in y])
```

```python
# inicializar los parametros
alpha = 0.01
b0 = 0
b1 = 0

# por cada batch
for epoch in range(1000):
    # calcular la prediccion
    y_hat = b0 + b1 * x_minmax

    # calcular el gradiente
    b0_grad = 2 * (y_minmax - y_hat).mean()
    b1_grad = 2 * ((y_minmax - y_hat)*x_minmax).mean()

    # actualizar los parametros
    b0 = b0 + alpha * b0_grad
    b1 = b1 + alpha * b1_grad

plt.scatter(x_minmax,y_minmax)
plt.plot(x_minmax,b0+b1*x_minmax)
plt.show()
    
print("FINAL:\n",b0,b1)
```
