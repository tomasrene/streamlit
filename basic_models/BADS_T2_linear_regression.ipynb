```python
import matplotlib.pyplot as plt

def linear_regression_analytical_sol(x,y):  
  # calculate the statistics
  x_mean = x.mean()
  y_mean = y.mean()

  # calculate b0 and b1
  b1 = sum([(x_i-x_mean) * (y_i-y_mean) for x_i,y_i in zip(x,y)])/sum([(x_i - x_mean)**2 for x_i in x])
  b0 = y_mean - b1 * x_mean

  # calculate final error
  error = sum([(y_i - (b0+x_i*b1))**2 for x_i,y_i in zip(x,y)])/len(x)

  # plot the results
  plt.scatter(x,y)
  plt.plot(x, b0 + b1 * x)
  plt.show()

  # return parameters and error
  return {'b0': b0,'b1':b1,'error':error}
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

def linear_regression_sklearn_sol(x,y):
  # convert to required format
  x = np.array(x).reshape(-1,1)
  y = np.array(y)

  # create the model
  lr = LinearRegression()

  # fit to data
  lr.fit(x,y)

  # calculate final error
  error = mean_squared_error(y, lr.predict(x))

  # plot the result
  plt.scatter(x,y)
  plt.plot(x, lr.intercept_ + lr.coef_[0] * x)
  plt.show()

  # return parameters and error
  return {'b0': lr.intercept_,'b1': lr.coef_[0],'error': error}
```

```python
import matplotlib.pyplot as plt

def lr_plot(x,y,b0,b1):
  plt.scatter(x,y)
  plt.plot(x, b0 + x*b1)
  plt.show()
  print()
```

```python
import numpy as np

def minmax(data):
  data_minmax = np.array([(datai - data.min())/(data.max() - data.min()) for datai in data])
  return data_minmax
```

```python
def linear_regression_gradient(x, y, alpha = 0.001, epochs=100):

  # random initialization of the parameters
  b0 = np.random.rand()
  b1 = np.random.rand()

  # normalize the variables
  x_minmax,y_minmax = minmax(x), minmax(y)

  # for each batch
  for epoch in range(epochs):
    # calculate the prediction
    y_hat = b0 + b1 * x_minmax

    # calculate the gradient
    b0_grad = 2 * (y_minmax - y_hat).mean()
    b1_grad = 2 * ((y_minmax - y_hat) * x_minmax).mean()

    # update the parameters
    b0 = b0 + alpha * b0_grad
    b1 = b1 + alpha * b1_grad

  # calculate final error
  error_final = ((y_minmax - (b0 + b1 * x_minmax))**2).mean()

  # plot the results
  lr_plot(x_minmax,y_minmax,b0,b1)
  
  return {'b0':b0,'b1':b1,'error':error_final}
```

```python
def linear_regression_sgd(x,y, alpha = 0.001, epochs=1000):

  # random initialization of the parameters
  b0 = np.random.rand()
  b1 = np.random.rand()

  # normalize the variables
  x_minmax,y_minmax = minmax(x), minmax(y)

  # for each batch
  for epoch in range(epochs):
    # for each element
    for i in range(len(x_minmax)):
      # calculate the prediction
      y_hat = b0 + b1 * x_minmax[i]

      # calculate the gradient
      b0_grad = 2 * (y_minmax[i] - y_hat).mean()
      b1_grad = 2 * ((y_minmax[i] - y_hat)*x_minmax[i]).mean()

      # update the parameters
      b0 = b0 + alpha * b0_grad
      b1 = b1 + alpha * b1_grad

  # calculate final error
  error_final = ((y_minmax - (b0 + b1 * x_minmax))**2).mean()

  # plot the results
  lr_plot(x_minmax,y_minmax,b0,b1)
  
  return {'b0':b0,'b1':b1,'error':error_final}
```

```python
def linear_regression_mbgd(x,y, alpha = 0.001, epochs=100, batch_size = 32):

  # random initiliazation of the parameters
  b0 = np.random.rand()
  b1 = np.random.rand()

  # normalization of variables
  x_minmax,y_minmax = minmax(x), minmax(y)

  # for each epoch
  for epoch in range(epochs):
    # calculate batches and apply to x and y
    indices = np.random.choice(len(x),batch_size)
    x_batch, y_batch = x_minmax[indices],y_minmax[indices]

    # calculate the prediction
    y_hat = b0 + b1 * x_batch

    # calculate the gradient
    b0_grad = 2 * (y_batch - y_hat).mean()
    b1_grad = 2 * ((y_batch - y_hat)*x_batch).mean()

    # update the parameters
    b0 = b0 + alpha * b0_grad
    b1 = b1 + alpha * b1_grad

  # calculate final error
  error_final = ((y_minmax - (b0 + b1 * x_minmax))**2).mean()
  
  # plot the results
  lr_plot(x_minmax,y_minmax,b0,b1)
  
  return {'b0':b0,'b1':b1,'error':error_final}
```

```python
import matplotlib.pyplot as plt

def lr_plot(x,y,b0,b1):
  plt.scatter(x,y)
  plt.plot(x, b0 + x*b1)
  plt.show()
  print()
```

```python
import numpy as np

def minmax(data):
  data_minmax = np.array([(datai - data.min())/(data.max() - data.min()) for datai in data])
  return data_minmax
```

```python
import keras

def linear_regression_keras(x,y,alpha=0.001,epochs=100,batch_size=32):
  
  # minmax
  x_minmax,y_minmax = minmax(x), minmax(y)

  # model
  modelo = keras.Sequential([keras.layers.Dense(
      1,
      input_shape=(1,),
      activation=None)
  ])

  # training
  modelo.compile(
      optimizer = keras.optimizers.SGD(learning_rate=alpha),
      loss = 'mse'
  )
  
  # fit
  modelo.fit(x_minmax,y_minmax,epochs=epochs,batch_size=batch_size, verbose=0)

  # coefficients
  b1,b0 = modelo.get_weights()[0][0,0],modelo.get_weights()[1][0]

  # final error
  error = modelo.history.history['loss'][-1]

  # plot
  lr_plot(x_minmax,y_minmax,b0,b1)

  return {'b0':b0,'b1':b1,'error':error}
```

```python
import time 

for df in [df_csv,df_bq]:
  start = time.time()
  print(linear_regression_keras(df.x, df.y,epochs=1000))
  end = time.time()
  print("Time: ",(end-start))
```
